```python
# Importe as bibliotecas necessárias
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error, mean_absolute_error
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
# Carregue um arquivo Excel em um DataFrame

# pelo caminho do seu arquivo de dados
data = pd.read_excel('/content/Treino (2).xlsx')

# Remova caracteres indesejados da coluna 'TEMP'
data['TEMP'] = data['TEMP'].str.replace("?", "")

# Remova o símbolo de porcentagem da coluna 'HUMIDADE'
data['HUMIDADE'] = data['HUMIDADE'].str.replace("%", "")

# Remova linhas com valores ausentes (NaN)
data = data.dropna()

# Converta os valores na coluna 'TEMP' para o tipo de dado float
data['TEMP'] = data['TEMP'].astype(float)

# Converta os valores na coluna 'HUMIDADE' para o tipo de dado float
data['HUMIDADE'] = data['HUMIDADE'].astype(float)

# O DataFrame 'data' agora contém os dados limpos e as colunas 'TEMP' e 'HUMIDADE'
# prontos para serem usados em análises ou modelagem de dados.
```

```python
X = data.drop('mg/dL', axis=1)
y = data['mg/dL']

rf = RandomForestRegressor(random_state=42)

# Treine o modelo
rf.fit(X, y)

# Obtenha as importâncias das features
importancias = rf.feature_importances_

# Crie um dataframe com as importâncias das features
df_importancias = pd.DataFrame({'feature': X.columns, 'importance': importancias})

# Ordene as importâncias em ordem decrescente
df_importancias = df_importancias.sort_values('importance', ascending=False)

# Selecione as 25 features com as maiores importâncias
top_features = df_importancias.head(22)['feature'].tolist()

# Crie um novo dataframe com apenas essas 30 colunas
df_novo = X[top_features]
df_novo
```

# testando sem normalização

```python
X = df_novo
y = data['mg/dL']

# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(df_novo, y, test_size=0.3, random_state=42, stratify=y)

# Definir os hiperparâmetros para cada modelo

rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

knn_params = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance']
}

gb_params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 1],
    'max_depth': [3, 5, 10]
}

# Definir os hiperparâmetros para os novos modelos
xgb_params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 1],
    'max_depth': [3, 5, 10]
}

lgbm_params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 1],
    'max_depth': [3, 5, 10]
}

# Definir a lista de modelos com seus respectivos hiperparâmetros
models = [
    (RandomForestRegressor(), rf_params),
 #   (KNeighborsRegressor(), knn_params),
 #   (GradientBoostingRegressor(), gb_params),
 #   (XGBRegressor(), xgb_params),
 #   (LGBMRegressor(), lgbm_params),
]

# Lista para armazenar os resultados
results = []

# Iterar sobre os modelos
for model, params in models:
    # Criar o objeto GridSearchCV
    grid_search = GridSearchCV(estimator=model, param_grid=params, cv=10, verbose=5,
                               return_train_score=True, scoring='neg_mean_squared_error')

    # Realizar o grid search com cross-validation
    grid_search.fit(X_train, y_train)

    # Obter o melhor modelo encontrado
    best_model = grid_search.best_estimator_

    # Fazer previsões com o melhor modelo
    y_pred = best_model.predict(X_test)

    # Calcular as métricas de erro
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    # Armazenar os resultados
    results.append({'Model': model.__class__.__name__,
                    'Best Parameters': grid_search.best_params_,
                    'RMSE': rmse,
                    'R²': r2,
                    'MAPE': mape,
                    'MAE': mae})

# Imprimir os resultados
for result in results:
    print('Model:', result['Model'])
    print('Best Parameters:', result['Best Parameters'])
    print('RMSE:', result['RMSE'])
    print('R²:', result['R²'])
    print('MAPE:', result['MAPE'])
    print('MAE:', result['MAE'])
    print()
```

# Esemble

```python
from sklearn.base import clone
from sklearn.ensemble import VotingRegressor
X = df_novo
y = data['mg/dL']

# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Definir os hiperparâmetros para cada modelo

rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

gb_params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 1],
    'max_depth': [3, 5, 10]
}


# Definir a lista de modelos com seus respectivos hiperparâmetros
models = [
    (RandomForestRegressor(), rf_params),
    (GradientBoostingRegressor(), gb_params),
]

# Lista para armazenar os resultados
results = []

# Iterar sobre os modelos
for model, params in models:
    # Criar o objeto GridSearchCV
    grid_search = GridSearchCV(estimator=model, param_grid=params, cv=10, verbose=5,
                               return_train_score=True, scoring='neg_mean_squared_error')

    # Realizar o grid search com cross-validation
    grid_search.fit(X_train, y_train)

    # Obter o melhor modelo encontrado
    best_model = grid_search.best_estimator_

    # Fazer previsões com o melhor modelo
    y_pred = best_model.predict(X_test)

    # Calcular as métricas de erro
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)

    # Armazenar os resultados
    results.append({'Model': model.__class__.__name__,
                    'Best Parameters': grid_search.best_params_,
                    'RMSE': rmse,
                    'R²': r2})

# Obter os melhores parâmetros para RandomForestRegressor e GradientBoostingRegressor
best_rf_params = None
best_gb_params = None
best_rmse = float('inf')
for result in results:
    if result['Model'] == 'RandomForestRegressor' and result['RMSE'] < best_rmse:
        best_rf_params = result['Best Parameters']
        best_rmse = result['RMSE']
    elif result['Model'] == 'GradientBoostingRegressor' and result['RMSE'] < best_rmse:
        best_gb_params = result['Best Parameters']
        best_rmse = result['RMSE']


# Criar o modelo VotingRegressor com RandomForestRegressor e GradientBoostingRegressor usando os melhores parâmetros
voting_model = VotingRegressor([
    ('random_forest', RandomForestRegressor(**best_rf_params)),
    ('gradient_boosting', GradientBoostingRegressor(**best_gb_params))

])

# Lista para armazenar os resultados do VotingRegressor
voting_results = []

# Treinar e testar o modelo VotingRegressor
voting_model.fit(X_train, y_train)
y_pred = voting_model.predict(X_test)

# Calcular as métricas de erro para o VotingRegressor
voting_rmse = mean_squared_error(y_test, y_pred, squared=False)
voting_r2 = r2_score(y_test, y_pred)

# Armazenar os resultados do VotingRegressor
voting_results.append({'Model': 'VotingRegressor',
                       'Best Parameters': None,
                       'RMSE': voting_rmse,
                       'R²': voting_r2})

# Adicionar os resultados do VotingRegressor aos resultados gerais
results.extend(voting_results)

# Imprimir os resultados
for result in results:
    print('Model:', result['Model'])
    print('Best Parameters:', result['Best Parameters'])
    print('RMSE:', result['RMSE'])
    print('R²:', result['R²'])
    print()
```
